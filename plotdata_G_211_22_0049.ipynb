```python
import matplotlib.pyplot as plt
import numpy as np

# Define axis limits and grid
plt.axis([0, 140, 0, 100])
plt.axis('on')
plt.grid(True)

# Data
t = [20, 40, 60, 80, 100, 120]  # Time
T = [30, 35, 43, 55, 70, 85]  # Temperature
p = [2, 3, 4, 5.3, 7.3, 9.6]  # Pressure
v = [0.6, 0.58, 0.54, 0.46, 0.35, 0.2]  # Volume

# Scale pressure and volume for plotting
pp = [i * 10 for i in p]  # Scale pressure by 10
vv = [i * 100 for i in v]  # Scale volume by 100

# Plot temperature, pressure, and volume
plt.plot(t, T, color='r', label='Temperature', marker='o')  # Plot temperature
plt.plot(t, pp, color='b', label='Pressure', marker='s')  # Plot scaled pressure
plt.plot(t, vv, color='g', label='Volume', marker='d')  # Plot scaled volume

# Add legend
plt.legend(loc='upper left')

# Plot pressure scale values
for y in np.arange(0, 100 + 1, 20):
    a = y / 10
    a = str(a)  # Convert to string for plotting as text
    plt.text(142, y, a, color='b')

# Label axes
plt.xlabel('Time (hrs)')
plt.ylabel('Temperature °K', color='r')

# Label pressure axis
plt.text(151, 65, 'Pressure (psi)', rotation=90, color='b')

# Plot volume scale values
for y in np.arange(100, -1, -20):
    a = y / 100
    a = str(a)
    plt.text(162, y, a, color='g')
    plt.text(159, y + 2, '_', color='g')

for y in np.arange(1, 99, 3):
    plt.text(157, y, '-', color='g')

# Label volume axis
plt.text(170, 65, 'Volume (cm^3)', rotation=90, color='g')

# Add title
plt.title('Compression Test Results')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
t = np.array([20, 40, 60, 80, 100, 120])  # Time
v = np.array([30, 35, 43, 55, 70, 85])   # Volume

# Calculate averages
c1 = np.mean(t)  # Average of t
c2 = np.mean(v)  # Average of v

# Calculate sums for regression
a = np.multiply(v, t)  # Element-wise multiplication of v and t
c3 = np.sum(a)         # Sum of a
a = np.multiply(t, t)  # Element-wise multiplication of t and t
c4 = np.sum(a)         # Sum of a

# Calculate line parameters A and B for linear regression
n = len(t)
A = (c3 - n * c1 * c2) / (c4 - n * c1 * c1)
B = c2 - A * c1

# Plotting the regression line and scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(t, v, color='b', label='Data Points')  # Scatter plot of data points
tp = np.arange(t[0], t[-1] + 2, 2)  # Generate points for plotting the line
vp = A * tp + B  # Calculate y values for the line
plt.plot(tp, vp, color='r', label='Regression Line')  # Plot the regression line

# Calculate and print RMS error
sumee = 0
for i in range(len(t)):
    ee = v[i] - (A * t[i] + B)
    sumee += ee ** 2

rms = np.sqrt(sumee / n)
print(f"Root Mean Square Error (RMS): {rms}")

# Plot settings
plt.xlabel('Time (hrs)')
plt.ylabel('Volume (cm^3)')
plt.title('Linear Regression: Volume vs Time')
plt.legend()
plt.grid(True)
plt.show()
```
